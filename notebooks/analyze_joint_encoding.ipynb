```python
import pandas as pd
from neural_priors.utils.data import get_all_subject_ids, Subject
import os.path as open
import os
import os.path as op

from tqdm.contrib.itertools import product
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
bids_folder = '/data/ds-neuralpriors'

subjects = get_all_subject_ids()
ranges = ['narrow', 'wide']

par_keys =  ['mu', 'r2', 'cvr2', 'sd', 'amplitude', 'baseline']
model_labels = [1,2,3,4,5]

df = []
keys= []

for subject_id, range_n, model_label in product(subjects, ranges, model_labels):
    
    try:
        sub = Subject(subject_id, bids_folder)
        tmp = sub.get_prf_parameters_volume(smoothed=True, joint=True, range_n=range_n, gaussian=True, roi='NPCr', keys=par_keys, model_label=model_label)
        df.append(tmp)
        keys.append((subject_id, range_n, model_label))
    except Exception:
        print(f'Erorr with {subject_id} {range_n} {model_label}')
        pass

df = pd.concat(df, keys=keys, names=['subject_id', 'range', 'model'])
```

```python
bids_folder = '/data/ds-neuralpriors'

subjects = get_all_subject_ids()
ranges = ['narrow', 'wide']

par_keys =  ['mu', 'sd', 'r2']

df = []
keys= []

for subject_id, range_n in product(subjects, ranges):
    try:
        sub = Subject(subject_id, bids_folder)
        tmp = sub.get_prf_parameters_volume(smoothed=True, joint=True, range_n=range_n, gaussian=True, roi='NPCr', keys=par_keys, model_label=2)
        df.append(tmp)
        keys.append((subject_id, range_n))
    except Exception as e:
        print(e)

df = pd.concat(df, keys=keys, names=['subject_id', 'range'])
```
