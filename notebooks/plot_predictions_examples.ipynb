```python
from neural_priors.utils.plotting import plot_prf_predictions
```

`plot_prf_predictions` offers an easy way to plot predictions over actual data. By default, it gets data from `NPCr` and takes the 12 voxels with the highest R2.

```python
plot_prf_predictions('17')
```

```python
plot_prf_predictions('12', smoothed=True)
```

We can also select specific voxels, for example because they have a weird shape

```python
from neural_priors.utils.data import Subject

subject = Subject('17', bids_folder='/data/ds-neuralpriors')
prf_pars = subject.get_prf_parameters_volume(None, cross_validated=False, smoothed=True, roi='NPCr', range_n='narrow', return_image=False)
```

```python
prf_pars[prf_pars['cvr2'] > 0.0].sort_values('fwhm')
```

```python
plot_prf_predictions('17', smoothed=True, roi='NPCr', range_n='narrow', voxels=[691, 323, 676, 559])
```

```python
prf_pars = subject.get_prf_parameters_volume(None, cross_validated=False, smoothed=True, roi='NPCr', range_n='wide', return_image=False)
prf_pars[prf_pars['cvr2'] > 0.0].sort_values('fwhm')
```

```python
plot_prf_predictions('17', smoothed=True, roi='NPCr', range_n='wide', voxels=[237, 293, 126])
```

```python
plot_prf_predictions('17', smoothed=True, roi='NPCr', range_n='both', voxels=[691, 323, 676, 559])
```
